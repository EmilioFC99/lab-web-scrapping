```python
url_rt = 'https://www.rottentomatoes.com/top/bestofrt/'

html_rt = requests.get(url_rt).content
soup_rt = BeautifulSoup(html_rt, 'lxml')
```

```python
print(soup_rt)
```

```python
tags = ['h1','h2','h3','p','a','href']

elements = [element.text for element in soup_rt.find_all(tags)]
```

```python
elements
```

```python
links = soup_rt.select('td a.unstyled.articleLink')
```

```python
links
```

```python
title_clean = [element.text.strip() for element in links]
```

```python
title_clean
```

```python
reviews = soup_rt.select('td[class="right hidden-xs"]')
reviews
```

```python
reviews_clean = [element.text for element in reviews]

reviews_clean
```

```python
url = 'https://www.rottentomatoes.com'
links[0]['href']
```

```python
links_to_movie = [url + links[i]['href'] for i in range(len(links))]
links_to_movie
```

```python
df = pd.DataFrame( list(zip(title_clean, reviews_clean, links_to_movie)), columns = ['Movie','Reviews', 'Link'] )
```

```python
df.sort_values('Reviews', ascending=True)
```
